# Final Report CS50AP Puyo

This notebook presents the results of the surveys made to the students at the end of the course. It is divided in two sections: one about the assignments and classes, and the other one about the instructor.

It is important to mention that eight students started the course, but later three abandoned it. When asked about the reasons they left the course, all three students stated lack of support from their parents. Their reasons were: "... my grades are low, so my parents made me left the course", "... I don't have support from my parents", "...my parents don't like that I spend too much time out of home attending the course". They said that the high school could help by explaining to their parents how important the course is for their future or including the course as part of the curriculum.

The data in this report is from the students that completed the course.

## Classes and Assignments

All students said to be satisfied with what they have learned in class, not only because of the specific programming knowledge but also because of the general tech literacy they have gained.

Let's see how difficult they found the classes:

## Conclusions

* Extracurricular activities are not a common thing in the local culture. This resulted in lack of support from the students' parents because they saw the course as not relevant enough and therefore thought that students shouldn't dedicate significant time and effort to the course. On the other hand, all students were aware of the importance of learning computer programming no matter what they later choose to study in university.

* It is fundamental to have support from the school authorities so that students who don't have their own laptop can use the computers in the laboratory. Also, the school can support with the events that the course promotes such as exposition of projects or future hackatons.

* The students often mentioned that they have too much homework and tests from their mandatory courses and they would like the school to reduce that load. Again, this alludes to the fact that extracurricular courses are not common. It would be helpful if the school includes the course as part of the curriculum.

* The material is ready for use in the next iteration of the course. This includes the slides for each class, all assignments, and a very useful [tool](https://github.com/cs50puyo/check50ap) created by the instructor to facilitate reviewing the assignments.
