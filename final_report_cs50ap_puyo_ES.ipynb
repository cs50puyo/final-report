```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
sv1 = pd.read_csv("Encuesta - Classes and Assignments.csv")
sv2 = pd.read_csv("Encuesta_ Instructor.csv")
data = pd.merge(sv1, sv2, on="Nombre de usuario")
```

# Reporte Final CS50AP Puyo

Este documento presenta los resultados de las encuestas realizadas a los estudiantes al final del curso. Está divido en dos secciones: la primera acerca de las clases y las tareas, y la segunda sección sobre el instructor.

El curso empezó con ocho estudiantes, pero luego tres de ellos se retiraron. Los estudiantes indicaron que la principal razón por la que dejaron el curso es la falta de apoyo de sus padres, mencionaron "...mis calificaciones han bajado, me han obligado a dejar el curso", "Falta de materiales, apoyo inexistente por parte de padres", "...no les gustaba la idea de que pase mucho tiempo en el curso (fuera de casa)". Los estudiantes indicaron que el colegio puede ayudar a que esto no siga pasando si es que le explican a sus padres que el curso es importante y que les servirá en el futuro, o incluyendo el curso como parte del currículum.

Los datos presentados en este reporte son de los estudiantes que completaron el curso.

## Clases y tareas

Todos los estudiantes indicaron estar satisfechos con lo que han aprendido en clase, no solamente por lo que han aprendido específicamente sobre programación sino también por lo que han aprendido acerca de tecnología en general.

Veamos cuán difíciles encontraron las clases.

```python
d = data["En general, ¿Qué tan difíciles son las clases?"].mean()/5*100
sizes = [50-(d/2), d/2, 50]
colors = ['silver', 'yellowgreen', 'white']

plt.pie(sizes, colors=colors, shadow=False)

centre_circle = plt.Circle((0,0),0.55,color='black', fc='white',linewidth=0)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.text(0.5,-0.15,'Muy difícil', fontsize=12)
plt.text(-1.05,-0.15,'Muy fácil', fontsize=12)
plt.text(-0.17, 0.65, str(d)+"%", fontsize=12)
plt.title("En general, ¿Qué tan difíciles son las clases?", fontsize=14, weight='bold')
plt.axis('equal')
plt.show()  
```

## Conclusiones

* Dado que las actividades extracurriculares no son parte de la cultura académica local, algunos padres de familia vieron al curso de programación como no lo suficientement importante como para que sus hijos dediquen tiempo y esfuerzo al mismo. Por otro lado, todos los estudiantes consideran importante aprender programación sin importar lo que luego decidan estudiar en la universidad.

* Es fundamental contar con el apoyo de las autoridades del colegio porque ellos permiten el acceso al laboratorio de computadores, que es primordial para los estudiantes que no tienen computador portátil, y porque el colegio puede ayudar con otros eventos que se promueven en el curso como la exposición de proyectos o futuros hackatons.

* Los estudiantes constantemente mencionaron que tiene demasiados deberes y pruebas en el colegio, y por tanto no tienen tiempo para el curso. Sería de enorme ayuda si el colegio puede incluir el curso de programación como parte del currículum.
